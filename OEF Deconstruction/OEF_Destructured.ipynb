# Destructuring $OEF

The final plot of first_day shows the intrinsic value of $OEF strictly based on the weighted value of its holdings

```python
import pandas as pd
import yfinance as yf
import numpy as np

tickers = pd.read_csv("OEF.csv")
tickers_list = tickers.values.tolist()
tickers = []
for x, i in enumerate(tickers_list):
    tickers.append(tickers_list[x][0])
prices = yf.download(tickers, period='5d', interval='1m')
print(prices)
```

```python
prices['Adj Close']
```

```python
mkt_cap = pd.DataFrame()
for i in tickers:
    ticker = yf.Ticker(i)
    share_count = ticker.get_shares_full(start='2021-01-01', end=None)
    caps = []
    for x in range(0, 1950):
        cap = share_count[-1] * prices['Adj Close'][i][x]
        caps.append(cap)
    mkt_cap[i] = caps

print(mkt_cap)
```

```python
total_cap = []
for i in range(0, 1950):
    total_cap.append(mkt_cap.iloc[i].sum())
total_cap
```

```python
weights = pd.DataFrame()
for b, i in enumerate(tickers):
    temp = []
    for x in range(0, 1950):
        temp.append(mkt_cap[i][x] / total_cap[b])
    weights[i] = temp

weights = weights.iloc[1: , :]
weights
```

```python
prices = prices['Adj Close']
prices
```

```python
ret = prices.pct_change()
ret = ret.reset_index()
ret = ret.iloc[1:, 1:]
ret.fillna(0)
ret
```

```python
IT_ret = ret.multiply(weights)
IT_ret
```

```python
IT_ret = IT_ret.sum(axis=1)
IT_ret
```
